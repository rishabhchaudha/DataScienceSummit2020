```python
import random
from numpy.random import choice
import pandas as pd
import math
import numpy as np
```

```python
class nqueens:    
    def __init__(self,Population=10,crossover1=0.3,crossover2=0.7,numberOfQueens = 8, generation = 10000):
        self.Population = Population
        self.crossover1 = crossover1
        self.crossover2 = crossover2
        self.numberOfQueens = numberOfQueens
        self.generation = generation
    
    #Fitnes Function
    def getFitness(pop):
        clashes = 0
        row = abs(len(set(pop)) - len(pop))
        clashes =  clashes + row
        for i in range(len(pop)):
            for j in range(len(pop)):
                if(i != j):
                    first = abs(i-j)
                    second = abs(pop[i] - pop[j])
                if(first == second):
                    clashes += 1
        return clashes
    
    #Generating Population
    def generatePopulation(population,numberOfQueens):
        n=numberOfQueens
        total_population = []
        for outloop in range(Population):
            random_data =  []
            for inloop in range(0,n):
                selected_data =  random.randint(0,n-1)
                while(selected_data in random_data):
                    selected_data =  random.randint(0,n-1)
                random_data.append(selected_data)
            total_population.append(random_data)
        clashes = []
        probability = []
        for outloop in range(Population):
            clashes.append(getFitness(total_population[outloop]))    
        probDataFrame =  pd.DataFrame({'Arrangement':total_population, 'clashes': clashes })
        probDataFrame = probDataFrame.sort_values('clashes').reset_index(drop = True)
        

        return probDataFrame,total_population
    
    #Drawing the board with queens
    def drawresult(elem):
        a = np.zeros((len(elem),len(elem)),dtype=str)
        b=elem
        for i in range(len(elem)):
            a[b[i]][i] = 'Q'
        print(a)
        
    def mutate(child):
        res = True
        while(res):
            randomInteger = random.randint(0,n-1)
            if(randomInteger not in child or len(set(child)) == n):
                res=False
        child[random.randint(0,n-1)] = randomInteger
        return child
    
    def crossoverAndMutate(crossover1,crossover2,elem1,elem2):
        crossOverPoint1 = int((n-1) * crossover1)
        crossOverPoint2 = int((n-1) * crossover2)
    
        child1 = elem1[0:crossOverPoint1]+elem2[crossOverPoint1:crossOverPoint2]+elem1[crossOverPoint2:]
        child2 = elem2[0:crossOverPoint1]+elem1[crossOverPoint1:crossOverPoint2]+elem2[crossOverPoint2:]    
        
        #Random Mutation
        child1[random.randint(0,n-1)] = random.randint(0,n-1)
        child2[random.randint(0,n-1)] = random.randint(0,n-1)
    
        #Controlled Mutation
        child1 = mutate(child1)
        child2 = mutate(child2)
    
        return child1,child2
    
    #Creating Generations
    def createGenerations(generation,probDataFrame,total_population,crossover1,crossover2):
        for outloop in range(generation):            
            #Selecting Two Chromosomes with highest Fitness
            draw=[]    
            draw.append(probDataFrame[0:1]['Arrangement'].values[0])
            draw.append(probDataFrame[1:2]['Arrangement'].values[0])
            
            if(getFitness(draw[0]) == 0 | getFitness(draw[1]) == 0):
                print(draw[0])
                print()
                drawresult(draw[0])
                break
            
            child1,child2 = crossoverAndMutate(crossover1,crossover2,draw[0],draw[1])
            total_population.append(child1)
            total_population.append(child2)
            
            fitness = []
    
            for inloop in range(len(total_population)):
                fitness.append(getFitness(total_population[inloop]))
    
    
            probDataFrame =  pd.DataFrame({'Arrangement':total_population, 'clashes': fitness})
            probDataFrame = probDataFrame.sort_values('clashes').reset_index(drop = True)
            
        return outloop
            
        
    if __name__=="__main__": 
        nq = nqueens()
        Population,crossover1,crossover1,numberOfQueens,generation = nq.Population,nq.crossover1,nq.crossover2,nq.numberOfQueens,nq.generation
        probDataFrame,total_population = generatePopulation(Population,numberOfQueens)
        
        outloop = createGenerations(generation,probDataFrame,total_population,crossover1,crossover2)
        print("Generation : "+ str(outloop))
        if(outloop == generation):
            print("Please try again")
        
```
